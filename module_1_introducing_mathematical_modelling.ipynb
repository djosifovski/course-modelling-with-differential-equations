# Selling Rainbowfish

## Problem statement

You want to breed rainbowfish to sell to pet stores. You start with a nice big aquarium and 30 fish, half of them male, half of them female. You want to predict the number of fish after a number of days, to see how many you can sell.

Let's define some quantities:

**Time**

We define $t$ as the time. The symbol for the unit of $t$ is $[t]$, and the most appropriate unit is days.

**Total number of rainbowfish**

We choose $P(t$) as the function which describes the number of rainbowfish at time $t$, where $P$ stands for population.

**Initial number of rainbowfish**

The initial population of rainbowfish is $P(0)=30$.

**Death rate**

Assume that a rainbowfish naturally lives for approximately 1000 days and that the ages of the fish in the aquarium are mixed. The death rate $d$ is the rate at which the fish die. (The unit $[d]$ of the death rate is per day.)
  - Rainbowfish live on average 1000 days (average livespan).
  - The death rate $d$ is defined as the number of deaths per unit of time per individual.
  - The death rate is inversely related to the average lifespan because if a fish lives for 1000 days on average, then, statistically, one fish is expected to die after 1000 days. Therefore, the rate at which fish die per day, per fish, can be expressed as the reciprocal of the average lifespan.
  - The death rate indicates the fraction of the fish population that dies each day relative to the total population. It indicates how many fish, on average, die daily per fish in the aquarium.. For instance, a death rate of 0.001 means that 0.1% of the fish population dies each day.
  - Expected Deaths: For a given population size, the death rate can be used to estimate the expected number of fish deaths per day. For example, if there are 1000 fish and the death rate is 0.001, we expect 1 fish to die each day.
  - Given the average lifespan of a rainbowfish is 1000 days, the death rate is $d = \dfrac{1}{1000} = 0.001$. This means that on average, 0.001 fish die each day per fish in the population, or 0.1% of the fish population dies each day.

**Birth rate**

Assume that one female rainbowfish lays eggs every 30 days and 42 of the eggs hatch to baby fish, half of them male, half female.

One female rainbowfish lays eggs every 30 days. Therefore, the average number of eggs laid per day by one female rainbowfish is $1/30$ eggs per day. On average one female rainbowfish produce $42/30 = 1.4$ baby rainbowfish per day.

The birth rate $b$ is the rate at which the rainbowfish replicate: the number of new fish *each* fish produces per day. The unit of $b$ is per day.

The birth rate $b$ specifically asks for the rate at which new fish (offspring) are produced. In this context, we're interested in the reproductive output in terms of offspring that can contribute to the population growth. While both male and female offspring are hatched, the birth rate typically focuses on the females because they are the ones that contribute directly to future reproduction and population growth. Therefore, $b$ reflects the rate at which female offspring are produced per day, as they play a direct role in the population dynamics.

We calculate $b$ based on the average number of female offspring produced per day by one female rainbowfish. Therefore, the value of the birth rate $b$ is then $21/30 = 0.7$ per day. This means that on average, each female rainbowfish produces offspring equivalent to 70% of its population size per day through reproduction.

## Mathematical model

**Balance equation:**

We have:
- $\Delta t$: lenght of a small time interval $[t, t + \Delta t]$
- Change in $P$ during $\Delta t$: $\Delta P = P(t + \Delta t) - P(t)$.

$
\begin{align}
\Delta P &= + \underbrace{bP(t)\Delta t}_{\text{Births}} - \underbrace{dP(t)\Delta t}_{\text{Deaths}} \\
&= + 0.7P(t)\Delta t - 0.001P(t)\Delta t \\
&= (0.7 - 0.001)P(t)\Delta t \quad \text{We can neglet the death rate because it is smaller in precision than the birth rate.}\\
&\approx 0.7 P(t)\Delta t \\
\end{align}
$

This is called a balance equation. The balance equation is valid for all times.

**Differential equation:**

We divide the balance equation with $\Delta t$:

$
\dfrac{\Delta P}{\Delta t} = 0.7 P(t)
$

We take the limit for $\Delta t$ to zero:

$\lim\limits_{\Delta t\rightarrow 0} \, \dfrac{\Delta P}{\Delta t} = \lim\limits_{\Delta t\rightarrow0} \,0.7\,P(t)$.

Thus, we get the differential equation which describes how the population changes:

$
\dfrac{dP}{dt} = 0.7 P(t)
$

with the initial condition $P(0) = 30$.

If you use only words to describe the differential equation you would say *"The derivative of the function equals a multiple of the function."* So the solution to the differential equation should be a function with this property.

The functions $f_1(t) = 0$ and $f_2(t) = c e^{kt}$ is the derivative a multiple of the same function, where $c$ and $k$ are constants. The derivative of an exponential is an exponential.

We have that $k = 0.7$ and $c = P(0) = 30$.

The form $P(t)=P(0)e^{0.7t}$ is obtained by integrating the differential equation $\frac{dP}{dt} = 0.7P(t)$, which leads to an exponential function due to the constant coefficient $0.7$. Multiplying by $P(0)$ ensures that the solution $P(t)$ satisfies the initial condition $P(0)$ at $t=0$. Thus, $P(0)$ acts as a scaling factor to adjust the amplitude of the exponential growth or decay based on the initial population size.

## Validation

```python
import numpy as np
import matplotlib.pyplot as plt
```

The graph of the solution $P(t)=30e^{0.7t}$ of the differential equation is:

```python
# Generate data
t = np.linspace(0, 10, 100)
P0 = 30  # Initial population
k = 0.7  # Growth rate

# Calculate population using the exponential growth formula
P = P0 * np.exp(k * t)

# Plotting
plt.figure(figsize=(6, 4))  # Adjust figure size
plt.plot(t, P, label='Population Growth', color='b', linewidth=2)  # Line color and width
plt.xlim([0, 5])  # Limit x-axis range
plt.ylim([0, 200])  # Limit y-axis range
plt.xlabel('Time (days)')  # Label x-axis
plt.ylabel('Population')  # Label y-axis
plt.title('Rainbowfish Population Growth')  # Plot title
plt.grid(True)  # Show grid
plt.legend(loc='upper left')  # Add legend
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

The solution is not realistic, because if you start with 30 rainbowfish, the population grows within two days to more than a hundred rainbowfish. This number keeps increasing, which means at some time the aquarium will be really full (and might even break). 

## A second modelling cycle for the rainbow fish

A mathematical model was formulated in the form of a differential equation and a solution was found. However, this solution is unrealistic: the number of rainbowfish grows very large very quickly.

In this section we will add one sentence to the problem, in the hope that this will limit the growth. We have chosen the following addition:

The aquarium owner expects to sell 20 rainbowfish every day.

## Mathematical model

The balance equation now becomes:

$
\Delta P = 0.7 P(t)\Delta t - 20 \Delta t
$

The differential equation is:

$
\dfrac{dP}{dt} = 0.7 P(t) - 20
$

### Calculation 1: Direction field & Equilibrium solution

A direction field for a differential equation provides a graphical representation of the slopes of solutions at various points in the $(t,P(t))$-plane. It consists of small line segments drawn in the $(t,P(t))$-plane, where each segment's slope corresponds to the value of $\frac{dP}{dt}$ at that point.

Purpose:
- Visualization: It helps visualize how solutions to the differential equation behave without solving it explicitly.
- Qualitative Analysis: By examining the direction field, one can understand the general behavior of solutions, such as where they increase, decrease, or remain constant.
- Initial Value Problem: It assists in identifying solutions that satisfy specific initial conditions $P(0)$, as these are trajectories that pass through the point $(0,P(0))$ in the direction field.

Interpretation: Trajectories of solutions will follow the direction indicated by these line segments. Solutions curve smoothly through the direction field, guided by the slopes provided.

Code: https://www.youtube.com/watch?v=Zobtt87M4G8

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation function
def dP_dt(t, P):
    return 0.7 * P - 20

# Define the range of P and t for the direction field
P_range = np.linspace(0, 100, 20)  # Population range from 0 to 100
t_range = np.linspace(0, 10, 20)   # Time range from 0 to 10

# Create a meshgrid of P and t values for the direction field
P, t = np.meshgrid(P_range, t_range)

# Calculate dP/dt for the meshgrid points
# Compute the slopes (dP/dt) at each point in the grid
dP = dP_dt(t, P)

# Normalize the dP values for better visualization
dP_normalized = dP / np.sqrt(1 + dP**2)

# Plotting the direction field and solutions together
plt.figure(figsize=(10, 8))

# Plot the direction field
plt.quiver(t, P, np.ones_like(dP_normalized), dP_normalized, scale=20, color='black')

# Add labels, title, legend, and grid
plt.xlabel('Time $t$')
plt.ylabel('Population $P$')
plt.title('Direction Field and Solutions of $\\frac{dP}{dt} = 0.7 P - 20$')
plt.xlim([0,10])
plt.ylim([0,60])
plt.grid(True)
plt.show()
```

A solution $P(t) = P_e$ which neither increases nor decreases, is an *equilibrium solution* of the differential equation. Because an equilibrium solution does not change, it has the property

$
\dfrac{dP}{dt}=0.
$

In other words, the equilibrium solution $P(t) = P_e$ is constant in time. Because it is constant, other words for an equilibrium solution are an *equilibrium point* and a *stationary point*.

We can sketch solutions to the differential equation if we know the initial condition:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation function
def dP_dt(t, P):
    return 0.7 * P - 20

# Define the range of P and t for the direction field
P_range = np.linspace(0, 100, 20)  # Population range from 0 to 100
t_range = np.linspace(0, 10, 20)   # Time range from 0 to 10

# Create a meshgrid of P and t values for the direction field
P, t = np.meshgrid(P_range, t_range)

# Calculate dP/dt for the meshgrid points
dP = dP_dt(t, P)

# Normalize the dP values for better visualization
dP_normalized = dP / np.sqrt(dP**2)

# Plotting the direction field and solutions together
plt.figure(figsize=(10, 8))

# Plot the direction field
plt.quiver(t, P, np.ones_like(dP_normalized), dP_normalized, scale=20, color='black', alpha=0.6)

# Define the initial conditions for the solutions
#initial_conditions = [10, 20, 27, 29, 31.58, 40, 50]  # Initial population sizes
initial_conditions = np.linspace(5,50,20)

# Time vector for solving the differential equation
t_span = (0, 10)  # Time span from 0 to 10
t_eval = np.linspace(0, 10, 100)  # Points at which to store the solution

# Plotting the solutions for each initial condition
for P0 in initial_conditions:
    # Solve the differential equation
    sol = solve_ivp(dP_dt, t_span, [P0], t_eval=t_eval)
    
    # Extract the solution
    t_sol = sol.t
    P_sol = sol.y[0]
    
    # Plot the numerical solution
    plt.plot(t_sol, P_sol, label=f'P(0) = {P0}', linewidth=2)

# Plot the steady state line
plt.axhline(20 / 0.7, color='red', linestyle='--', label='P(0) = 28.57 (Steady State)')

# Add labels, title, legend, and grid
plt.xlabel('Time $t$')
plt.ylabel('Population $P$')
plt.title('Direction Field and Solutions of $\\frac{dP}{dt} = 0.7 P - 20$')
#plt.legend()
plt.xlim([0,10])
plt.ylim([0,60])
plt.grid(True)
plt.show()
```

### Calculation 2: Phase line & Stability

Looking at the direction field we see two regions in which the solution curves behave differently. If the initial value is below the equilibrium, the solution curve moves away from the equilibrium and towards minus infinity as time increases, and if the initial value is above the equilibrium, the solution curve moves away from the equilibrium and tends to infinity as time increases.

You can only draw a phase line when the differential equation is a so-called autonomous differential equation.

In an autonomous differential equation, none of the terms depend explicitly on the independent variable.

A differential equation that is not autonomous is for example $\dfrac{dy}{dt} = 5y(t) +\sin(t)$. For this equation you cannot draw a phase line.

A phase line is a graphical representation used to analyze the qualitative behavior of a first-order differential equation, particularly those of the form $\dfrac{dP}{dt} = f(P)$, where $P$ represents a population or some quantity changing over time $t$. The phase line plots the derivative $\dfrac{dP}{dt}$​ as a function of $P$, indicating whether $P$ increases, decreases, or remains constant at different values of $P$. This function determines the slope of the phase line at each point $P$.

```python
# Define the differential equation function
def dP_dt(P):
    return 0.7 * P - 20

# Define the range of P values
P_values = np.linspace(0, 100, 200)  # Adjust the range and number of points as needed

# Calculate dP/dt for the range of P values
dP_values = dP_dt(P_values)

# Plot the phase line
plt.figure(figsize=(8, 6))
plt.plot(P_values, dP_values, label='$\\frac{dP}{dt} = 0.7P - 20$', color='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)  # Zero line for reference
plt.xlabel('Population $P$')
plt.ylabel('$\\frac{dP}{dt}$')
plt.title('Phase Line for $\\frac{dP}{dt} = 0.7P - 20$')
plt.grid(True)
plt.legend()
plt.show()
```

In this graph you can see that if $P(t)$ is smaller than the equilibrium value $20/0.7$, the rate of change $\dfrac{dP}{dt}$ is negative, so the value of $P(t)$ will decrease. For values of $P(t)$ larger than the equilibrium, the value of $\dfrac{dP}{dt}$ is positive

**Stable equilibrium point:**

We call an equilibrium point stable if any initial value close to the equilibrium point gives solutions that always remain close to the equilibrium point.

For many stable equilibrium points, the solutions that start close by, do not only remain close by, but for $t \rightarrow \infty$ the solutions even tend to the equilibrium point.

**Unstable equilibrium point:**

Any equilibrium point which is not stable we call unstable, so there is at least one initial value close to the equilibrium which will give a solution that moves away from the equilibrium point.

The equilibrium solution $P_e = \dfrac{20}{0.7}$ is an unstable equilibrium solution.
